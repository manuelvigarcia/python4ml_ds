### Parsing dates at read time

```python
sales = pd.read_csv('../jupyter_notebooks/03-Pandas/RetailSales_BeerWineLiquor.csv', parse_dates=[0])
```

```python
sales['DATE']
```

```python
sales = sales.set_index("DATE")
```

```python
sales
```

```python
# Now we can use "resample()" to aggregate data by time units according to different rules.
# 'YE' is "end of year" (https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects)
sales.resample(rule='YE').mean()
```

### Accessing datetime objects using the .dt method
All available features for a single python datetime object are also available Series- and Dataframe-wise by using the .dt. method

```python
sales = pd.read_csv('../jupyter_notebooks/03-Pandas/RetailSales_BeerWineLiquor.csv', parse_dates=[0])
sales
```

```python
sales.info()
```

```python
sales['DATE'].dt.year
```

```python
sales['DATE'].dt.month
```

```python
sales['DATE'].dt.day
# Seems all data points are at the beginning of the month
```

```python
sales['DATE'].dt.hour
# no time of day information
```
