# Missing Data
You may:
- continue with missing data
- Remove the entire observation
- Replace with a sensible substitution

```python
import numpy as np
import pandas as pd
```

NaN is a value that is not present.

It will not be equal to anything, since it is not known.

To test if a value is NaN, you shall use "is np.nan" and not "== np.nan"

```python
df = pd.read_csv('../jupyter_notebooks/03-Pandas/movie_scores.csv')
df
```

```python
#check where there are null and not null values.
df.isnull()
```

```python
df.notnull()
```

```python
# filter by that filter
df[df['pre_movie_score'].notnull()]
```

## Three options with missing data
 1. keep the data - Just keep going, keeping in mind that you miss some values and which ones.
 2. drop the data - dropna
 3. fill the data

### Drop the data

```python
help(df.dropna)
```

```python
#drop any row having any NaN
df.dropna()
```

```python
# drop all rows that have less than 1 non null value
df.dropna(thresh=1)
```

```python
# if you ask for 5 non null values, Hugh gets dropped too
df.dropna(thresh=5)
```

```python
# rows versus column focus
#   the default behaviour: drop rows with NaN values
df.dropna(axis=0)
```

```python
#   going for columns: drop any column containing any NaN value
df.dropna(axis=1)
```

. . . because we do have a column with all values as NaN

You can drop rows that miss critical values only. Use the 'subset' parameter.

```python
#Define which columns are the important ones and drop rows that miss any of them.
df.dropna(subset=['first_name', 'last_name', 'age', 'sex'])
```

## Fill in the data
This is the most difficult, since it involves knowing the domain in order not to bias the data

Use 'fillna()' to provide substitutes for NaN.

```python
help(df.fillna)
```

```python
df.fillna('Missing')
```

```python
# but it is better to fill by column, or at least by type of value
df[['pre_movie_score','post_movie_score']].fillna(0.0)
```

```python
# to place it in the dataframe and make the change permanent, you have to assign it.
df['pre_movie_score'] = df['pre_movie_score'].fillna(0.0)
df['post_movie_score'] = df['post_movie_score'].fillna(0.0)
df
```

```python
# But it is common to use a function (e.g. mean()) to fill the values so the filled in will not deviate the existing data
# Let's reset the dataframe and do that
df = pd.read_csv('../jupyter_notebooks/03-Pandas/movie_scores.csv')
df
```

```python
df['pre_movie_score'] = df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
```

```python
df
```

```python
# if you are OK with filling all numeric values with the mean, you can just use it "generically"
df.fillna(df.select_dtypes(include=['number']).mean())
```
