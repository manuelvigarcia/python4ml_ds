## Low level Figures and Axes

```python
# create data
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
#Figure and all plotting in the same cell
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y);
```

```python
# same for a and b data
#Figure and all plotting in the same cell
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
```

```python
#Figure with two sets of axes at different sizes.
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);

axes2 = fig.add_axes([0.2,0.2,0.1,0.1])
```

```python
#Figure with two sets of axes at different sizes.
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

axes2 = fig.add_axes([0.2,0.2,0.25,0.25])
axes2 = plt.plot(x,y)
```

```python
#Figure with two sets of axes. Make one a Zoom-in section of the other
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

axes2 = fig.add_axes([0.2,0.2,0.25,0.25])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,20)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')
```

```python
fig = plt.figure(figsize=(8,5),dpi=100)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

## subplots

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y)
axes[0,1].plot(a,b)

plt.tight_layout() # Auto-space to fit all the graph elements
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y)
axes[0,1].plot(a,b)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)
# Manual spacing by % of an axe
fig.subplots_adjust(wspace=0.6,hspace=0.3)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y)
axes[0,1].plot(a,b)
axes[1,0].plot(x,y)
axes[1,0].set_ylabel('Y Label 1,0')
axes[1,0].set_xlabel('X Label 1,0')

axes[1,1].plot(a,b)
axes[1,1].set_ylabel('Y Label 1,1')
axes[1,1].set_xlabel('X Label 1,1')

fig.suptitle("Multiple plots")
plt.tight_layout()

fig.savefig('MultiplePlots.png', bbox_inches='tight')
```

## Styling: colors and styles

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color='purple')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color='purple')
ax.plot(x,x+1, color= '#e6c017')
```

```python
# line width

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color='purple', lw=10)
ax.plot(x,x+1, color= '#e6c017', linewidth=0.5)
```

### line style

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# possible linestyle options: '--', '-','-.',':','steps'
ax.plot(x,x, color='purple', lw=10,ls='--')
ax.plot(x,x+1, color= '#e6c017', linewidth=5, linestyle='-.')
ax.plot(x,x+2, color='blue', lw=5,ls='-.')
ax.plot(x,x+3, color='gray', lw=5,ls='-')
ax.plot(x,x+4, color='orange', lw=5,ls=':')
# to set a custom dot pattern line
lines = ax.plot(x,x+5, color='red', lw=7,ls='-') # we are going to substitute the '-' with custom dashed style
lines[0].set_dashes([1,2,1,2,5,2]) #pairs of 'line dots, space dots'
lines = ax.plot(x,x+6, color='darkgreen', lw=7,ls='-')
lines[0].set_dashes([1,2,2,2,3,2,2,2]) #pairs of 'line dots, space dots'
```

### markers

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# possible linestyle options: '--', '-','-.',':','steps'
ax.plot(x,x, color='purple', lw=1,ls='--', marker='o')
ax.plot(x,x+1, color= '#e6c017', linewidth=1, linestyle='-.', marker='>')
ax.plot(x,x+2, color='blue', lw=1,ls='-.', marker='v', markersize=20, markerfacecolor='cyan',markeredgewidth=5,markeredgecolor='lightblue')
ax.plot(x,x+3, color='gray', lw=1,ls='-', marker='x',ms=15)
ax.plot(x,x+4, color='orange', lw=1,ls=':', marker='X')
# to set a custom dot pattern line
lines = ax.plot(x,x+5, color='red', lw=1,ls='-', marker='.') # we are going to substitute the '-' with custom dashed style
lines[0].set_dashes([1,2,1,2,5,2]) #pairs of 'line dots, space dots'
lines = ax.plot(x,x+6, color='darkgreen', lw=1,ls='-', marker='P')
lines[0].set_dashes([1,2,2,2,3,2,2,2]) #pairs of 'line dots, space dots'
```
