## apply() with multiple columns and lambdas

```python
def tip_quality(total_bill,tip):
    if (tip/total_bill > 0.25):
        return "Generous"
    else:
        return "Other"
```

```python
tip_quality(16.99, 1.01)
```

```python
df.apply(lambda df: tip_quality(df['total_bill'],df['tip']),axis=1)
```

```python
df['tip_quality']=df[['total_bill','tip']].apply(lambda df: tip_quality(df['total_bill'],df['tip']),axis=1)
df.head()
```

### using numpy vectorize

```python
df['tip_quality'] = np.vectorize(tip_quality)(df['total_bill'],df['tip'])
df.head()
```

```python
df.tail()
```

```python
df[df['tip_quality'] == 'Generous']
```

```python
import timeit
```

```python
setup = '''
import pandas as pd
import numpy as np
df = pd.read_csv('../jupyter_notebooks/03-Pandas/tips.csv')
def tip_quality(total_bill,tip):
    if (tip/total_bill > 0.25):
        return "Generous"
    else:
        return "Other"
'''
```

```python
stmt_one= '''
df['tip_quality']=df[['total_bill','tip']].apply(lambda df: tip_quality(df['total_bill'],df['tip']),axis=1)
'''
```

```python
stmt_two='''
df['tip_quality'] = np.vectorize(tip_quality)(df['total_bill'],df['tip'])
'''
```

```python
timeit.timeit(setup=setup, stmt=stmt_two,number=1000)
```

```python
timeit.timeit(setup=setup, stmt=stmt_one,number=1000)
```
