# Pandas Pivot Tables
Tehy reorganise information in a DataFrame, they do not find out or calculate new information: they just reorgainise, focus, and, maybe, drop some information.

It only makes sense to pivot DataFrames with repeated values over each column of interest.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../jupyter_notebooks/03-Pandas/Sales_Funnel_CRM.csv')
```

```python
df
```

```python
# How many licenses of each type did Google purchase?
licenses =df[['Company','Product','Licenses']]
licenses
```

```python
pd.pivot(data=licenses, index='Company',columns='Product',values='Licenses')
```

```python
#How much have we sold per company
pd.pivot_table(df,index="Company",aggfunc='sum')
```

```python
#You can achieve the same with groupby
df.groupby('Company').sum()
```

```python
# To limit the columns that are operated, you may use values argument
pd.pivot_table(df,index="Company",aggfunc='sum',values=['Licenses','Sale Price'])
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],aggfunc='sum')
```

```python
# Add in information about products
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc='sum')
```

```python
# do something about the NaN values
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc='sum',fill_value=0)
```

```python
# if you use numpy functions for the aggregation, you can specify more than one
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc=[np.sum,np.mean],fill_value=0)
```

```python
# Play around whith what are columns and rows
pd.pivot_table(df,index=['Account Manager','Contact','Product'],values=['Sale Price'],aggfunc=[np.sum,np.mean],fill_value=0)
```

```python
#Use margins to sumarize
pd.pivot_table(df,index=['Account Manager','Contact','Product'],values=['Sale Price'],aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
```
