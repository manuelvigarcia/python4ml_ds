## Operations

```python
#imaginary sales data for two quarters of a company
q1 = {'Japan':80, 'China':450, 'India':200, 'USA':250}
q2 = {'Brazil':100,'China':500,'India':210, 'USA':260}
```

```python
sales_q1 = pd.Series(q1)
sales_q1
```

```python
sales_2 = pd.Series(q2)
sales_2
```

```python
sales_q1['Japan']
```

```python
sales_q1.keys()
```

```python
sales_q1 *2
```

```python
# just sum both will not know how to deal with missing values
sales_q1 + sales_2
```

```python
# You can use np.sum() which has an argument to use when the value is missing
sales_q1.add(sales_2,fill_value=0)
```

```python
sales_q1
```
