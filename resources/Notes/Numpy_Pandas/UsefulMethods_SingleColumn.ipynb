# Useful Methods to apply in a single column

```python
import pandas as pd
import numpy as np
```

# apply()

```python
df = pd.read_csv('../jupyter_notebooks/03-Pandas/tips.csv')
df
```

```python
#last 4 digits of the credit card
#    to get the last 4 characters of a string: 
'this is a text string'[-4:]
```

```python
#     to change 'CC number' into string
str(df['CC Number'])
```

```python
# not quite, so build a custom function; then apply() it to the column
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(1234567890)
```

```python
#   do the magic
df['CC Number'].apply(last_four)
```

```python
# save it as a column
df['last_four'] = df['CC Number'].apply(last_four)
df.head()
```

```python
# qualify the restaurant as "$", "$$", "$$$"
def yelp(price):
    if price < 10:
        return "$"
    elif price < 20:
        return "$$"
    else:
        return "$$$"
```

```python
yelp(5)
```

```python
yelp(15)
```

```python
yelp(25)
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
df
```

```python
# try operating more than one column. Create the 'tip%' column.
```

```python
def tip_percentage(total, tip):
    return round(tip/total*100,2)
```

```python
tip_percentage(16.99, 1.01)
```

```python
df['tip%'] = (df['total_bill'],df['tip']).apply(tip_percentage)
```
