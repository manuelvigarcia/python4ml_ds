In case you face something difficult to perform in one step, you may always build a function and use the apply method with it

```python
def cleanup(name):
    name = name.replace(';','')
    name = name.strip()
    name = name.capitalize()
    return name
```

```python
messy_names
```

```python
messy_names.apply(cleanup)
```
