# groupby() operations
groupby() allows to examine data per category

groupby() is lazy: it does not evaluate until there is something to do with it.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../jupyter_notebooks/03-Pandas/mpg.csv')
df
```

```python
# let's see what can be a category
for i in df.columns.values:
    print(df[i].unique())
```

So cylinders, year and name seem to be practical categories.

```python
df[['mpg', 'cylinders', 'displacement', 'weight', 'acceleration', 'model_year', 'origin']].groupby('model_year').mean()
```

```python
# Group by several columns
```

```python
df[['mpg', 'cylinders', 'displacement', 'weight', 'acceleration', 'model_year', 'origin']].groupby(['model_year', 'cylinders']).mean()
```

This is a multilevel index dataframe. For each year, it has all the existing numbers of cylinders made that year, and for each, the mean of all the other columns is provided.

```python
year_cyl = df[['mpg', 'cylinders', 'displacement', 'weight', 'acceleration', 'model_year', 'origin']].groupby(['model_year', 'cylinders']).mean()
year_cyl
```

From this multilevel index dataframe you can access any particular piece. Let's start by seeing how the index look like.

```python
year_cyl.index
```

```python
year_cyl.index.levels
```

```python
year_cyl.loc[82]
```

```python
# and for only the 6 cylinders for year 82 you can use the MultiIndex tuple
year_cyl.loc[(82,6)]
```

```python
# To get a cross section of the dataframe, e.g. all 4 cylinder data but for every year.
year_cyl.xs(key=4, level='cylinders')
```

You can also aggregate and apply different aggregation functions to different columns of a dataframe.

Let's see how to get the maximum value and the mean value for the 'mpg' column, and the mean and standard deviation for the weigth from the original dataframe

```python
df
```

```python
df.select_dtypes(include=['number']).agg(['std','mean'])
```

To apply different aggregation functions to different columns, you may pass a dictionary of functions by columns

```python
df.agg({'mpg':['max', 'mean'],'weight':['mean','std']})
```
