## Conditional filtering Multiple conditions

```python
# using AND & and OR | operators
```

```python
# total bill > 30
df['total_bill'] > 30
```

```python
# males
df['sex'] == 'Male'
```

```python
# filter by both conditions
df[(df['total_bill'] > 30) & (df['sex'] == 'Male')]
```

## Conditional filtering multiple values and categories

```python
# bills on weekend
df[(df['day'] == 'Sun') | (df['day'] =='Sat')]
```

```python
options = ['Sat','Sun']
```

```python
df[df['day'].isin(options)]
```

```python
df[df['day'].isin(['Sat','Sun'])]
```
