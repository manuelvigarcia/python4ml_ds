# Distribution plots

present continuous data.

* Rug plot
* Histogram
* KDE (Kernel Distribution Estimation) Plot

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../jupyter_notebooks/05-Seaborn/dm_office_sales.csv')
df
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.displot(data=df,x='salary')
```

```python
sns.displot(data=df,x='salary',bins=10)
```

```python
sns.displot(data=df,x='salary',bins=100)
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary')
```

```python
sns.displot(data=df,x='salary',bins=20,color='red',edgecolor='blue', linewidth=3,linestyle='--')
```

```python
sns.histplot(data=df,x='salary')
```

```python
# show all three: histogram, rug, KDE
sns.displot(data=df,x='salary',kde=True,rug=True)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
sns.displot(data=sample_ages,x='age',rug=True,bins=30)
```

```python
sns.displot(data=sample_ages,x='age',rug=True,bins=30,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x='age')
```

```python
#Since there are no people with negative age
sns.kdeplot(data=sample_ages,x='age',clip=[0,150])
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,150],bw_adjust=0.6)
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,150],bw_adjust=0.6,fill=True)
```
