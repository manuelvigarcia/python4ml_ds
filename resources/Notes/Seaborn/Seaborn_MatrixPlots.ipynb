# Matrix Plots

* Heatmaps
* Clustermaps

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../jupyter_notebooks/05-Seaborn/country_table.csv')
df
```

```python
df = df.set_index('Countries')
```

```python
df
```

```python
sns.heatmap(df)
```

'Life expectancy' is breaking the heatmap because it is in a different unit. You may change it to a rate or drop it alltogether.

```python
sns.heatmap(df.drop('Life expectancy',axis=1))
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5)
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5, annot=True)
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5, annot=True, cmap='viridis')
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5, annot=True, cmap='viridis',center=30)
```

```python
sns.clustermap(df.drop('Life expectancy',axis=1),linewidth=0.5, annot=True, cmap='viridis')
```

```python
sns.clustermap(df.drop('Life expectancy',axis=1),linewidth=0.5, annot=True, cmap='viridis',col_cluster=False)
```
