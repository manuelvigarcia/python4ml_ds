# Comparison plots
These are plots aimed at comparing two metrics or two dimensions of a dataset.

* Jointplot - scaterplot(or hexplot) + histogram (or KDE plot)
* Pairplot - Compare all numerical columns of a dataframe, can be CPU-intensive with more columns. It consists of a matrix of scatterplots or KDE plots, with the diagonal having KDE plots or histograms

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../jupyter_notebooks/05-Seaborn/StudentsPerformance.csv')
df
```

## Jointplot

```python
sns.jointplot(data=df,x='math score', y='reading score')
```

```python
sns.jointplot(data=df,x='math score', y='reading score',alpha=0.4)
```

```python
sns.jointplot(data=df,x='math score', y='reading score',alpha=0.4,size=2)
```

```python
sns.jointplot(data=df,x='math score', y='reading score',kind='hex')
```

```python
sns.jointplot(data=df,x='math score', y='reading score',kind='hist')
```

```python
sns.jointplot(data=df,x='math score', y='reading score',kind='kde')
```

```python
sns.jointplot(data=df,x='math score', y='reading score',kind='kde',fill=True)
```

```python
sns.jointplot(data=df,x='math score', y='reading score',alpha=0.4,size=2, hue='gender')
```

## Pair plot

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='gender')
```

```python
sns.pairplot(data=df,hue='gender',kind='hist')
```

```python
# It is symetrical across the variable, so we can ommit the upper half.
sns.pairplot(data=df,hue='gender',corner=True)
```
