# Categorical Plots

There are specific kinds of plots better suitted to represent categorical data (not continuous)
* countplot
* barplot

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../jupyter_notebooks/05-Seaborn/dm_office_sales.csv')
df
```

```python
df.info()
```

```python
df['division'].value_counts()
```

```python
sns.countplot(data=df,x='division',hue='division')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df,x='division',hue='division')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df,x='division',hue='division')
plt.ylim(0,1000)
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df,x='level of education',hue='division')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df,x='level of education',hue='division',palette='Set2')
```

```python
sns.barplot(data=df,x='level of education',y='salary',hue='level of education',estimator=np.mean,errorbar='sd')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='level of education',y='salary',hue='division',estimator=np.mean,errorbar='sd')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='level of education',y='salary',hue='division',estimator=np.mean,errorbar='sd')
plt.legend(bbox_to_anchor=(1.05,1))
```

## Box plot
Displays distribution of a continuous variable.

It uses quartiles for that

```python
sns.boxplot(data=df,x="salary")
```

## Violin plot
Shows distribution by placing 2 KDE plots mirroring each other, and with the boxplot information inside: dot showing the mean and a "box" showing the IQR.

```python
sns.violinplot(data=df, x='division', y='salary',hue='division')
```

```python
sns.violinplot(data=df, x='level of education', y='salary',hue='level of education')
```

## swarmplot
Tries to show all data points in a KDE fashion.

```python
sns.swarmplot(data=df,x='salary')
```

```python
plt.figure(figsize=(12,5))
sns.swarmplot(data=df,x='division',y='salary',hue='division')
```

## Boxenplot or Letter-value plot
This is an improvement on the boxplot.

Only use it if the audience is familiar with it.

It uses quantiles instead of quartiles and it bridges the gap between KDE and boxplot

```python
sns.boxenplot(data=df,x='division',y='salary',hue='division')
```

Now, play with what is shown inside the violin.

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=df,x='reading score', y='parental level of education',hue='test preparation course',split=True,inner=None)
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=df,x='reading score', y='parental level of education',hue='test preparation course',split=True,inner='quartile')
```

```python
# One line for each observation
plt.figure(figsize=(8,6))
sns.violinplot(data=df,x='reading score', y='parental level of education',hue='test preparation course',split=True,inner='stick')
```

Now play with how close the contour resembles the actual data distribution: bandwidth parameter

```python
# One line for each observation
plt.figure(figsize=(8,6))
sns.violinplot(data=df,x='reading score', y='parental level of education',hue='test preparation course',split=True,bw_method=0.2)
```

```python
sns.swarmplot(data=df, x='math score')
```

Avoid the warning about the % of data not placed by enlarging the figure or reducing the dot size
