# Seaborn Grids
Grids of plots 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../jupyter_notebooks/05-Seaborn/StudentsPerformance.csv')
df
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch',row='test preparation course')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch',row='test preparation course',hue='race/ethnicity')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='violin',col='lunch',row='test preparation course',hue='race/ethnicity')
```

```python
sns.pairplot(df)
```

```python
g=sns.PairGrid(df)
g=g.map_upper(sns.scatterplot)
g=g.map_lower(sns.kdeplot)
g=g.map_diag(sns.histplot)
```

```python
g=sns.PairGrid(df, hue='gender')
g=g.map_upper(sns.scatterplot,alpha=0.4,size=2)
g=g.map_lower(sns.kdeplot,fill=True)
g=g.map_diag(sns.histplot)
g=g.add_legend()
```
