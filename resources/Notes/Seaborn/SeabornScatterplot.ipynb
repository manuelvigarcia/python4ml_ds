# Scatter plots

Present continuous (not categorical) features.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../jupyter_notebooks/05-Seaborn/dm_office_sales.csv')
df
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales', data=df,hue='salary')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales', data=df,hue='level of education')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales', data=df,hue='level of education',palette='Dark2')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales', data=df,hue='level of education',palette='Dark2', size = 'salary')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales', data=df,hue='level of education',palette='Dark2', size = 'salary', alpha=0.4)
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales', data=df,style='level of education', hue='level of education', size = 'salary', alpha=0.6)
plt.savefig('../Notes/Seaborn/scatter_plot_test.jpg')
```
